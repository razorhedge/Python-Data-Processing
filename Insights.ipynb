# Carga de Datos

```python
mainpath = "C:/Users/mromo/Desktop/Rappi/Datasets/S"+semana1+"/Data"
mainpath2 = "C:/Users/mromo/Desktop/Rappi/Datasets/S"+semana2+"/Data"
mainpath3 = "C:/Users/mromo/Desktop/Rappi/Datasets/BD Datos RT.xlsx"
conexion_path = "Info Conexion.csv"
fp1 = os.path.join(mainpath,conexion_path)
fp2 = os.path.join(mainpath2,conexion_path)
```

# Data Frames

```python
df1 = pd.read_csv(fp1, parse_dates = ['fecha'])
df2 = pd.read_csv(fp2, parse_dates = ['fecha'])
df3 = pd.read_excel(mainpath3, parse_dates = ['created_at'])
```

```python
df1.head()
```

```python
df_nuevos = df3[df3['created_at']>='2018-10-10']
nuevos_id = df_nuevos['id'].tolist()
```

df1group = df1.groupby('skid').sum()
df2group = df2.groupby('skid').sum()

```python
df1group = df1group.reset_index()
df2group = df2group.reset_index()
```

```python
df3 = pd.merge(df1group, df2group, on='skid', how='outer')
```

```python
df3.head()
```

```python
df3 = df3.drop(['hora_x', 'hora_y'], axis=1)
```

```python
df3 = df3.rename(columns = {'time_x':'Horas S'+semana1, 'orders_x':'Ordenes S'+semana1, 'ganancia_x':'Ganancia S'+semana1,
                      'time_y':'Horas S'+semana2, 'orders_y':'Ordenes S'+semana2, 'ganancia_y':'Ganancia S'+semana2})
df3['Horas S'+semana1] = df3['Horas S'+semana1].apply(lambda x: x/60)
df3['Horas S'+semana2] = df3['Horas S'+semana2].apply(lambda x: x/60)
```

```python
df6 = df3
df6.insert(7, 'Variacion en Horas de Conexion', np.abs(df6['Horas S'+semana1]-df6['Horas S'+semana2]))
```

```python
df6
```

```python
df3['Horas S'+semana1] = df3['Horas S'+semana1].fillna(df3['Horas S'+semana1].mean())
df3['Horas S'+semana2] = df3['Horas S'+semana2].fillna(df3['Horas S'+semana2].mean())
df3['Ordenes S'+semana1] = df3['Ordenes S'+semana1].fillna(df3['Ordenes S'+semana1].mean())
df3['Ordenes S'+semana2] = df3['Ordenes S'+semana2].fillna(df3['Ordenes S'+semana2].mean())
df3['Ganancia S'+semana1] = df3['Ganancia S'+semana1].fillna(df3['Ganancia S'+semana1].mean())
df3['Ganancia S'+semana2] = df3['Ganancia S'+semana2].fillna(df3['Ganancia S'+semana2].mean())
```

```python
df_data = df3.describe().astype(int)
df_data = df_data.rename(index = {'count':'Total RTs Conectados', 'mean':'Tiempo Promedio Conexion'})
```

```python
df4 = df3[df3['Horas S'+semana2]>40].sort_values(['Ordenes S'+semana2,'Horas S'+semana1], ascending = False)
df4 = df4.fillna(0)
df5 = df4[df4['Horas S'+semana1]<6]
```

```python
df5.shape
```

```python
writer = pd.ExcelWriter('C:/Users/mromo/Desktop/Rappi/Datasets/Insights S'+semana2+'.xlsx')
df6.to_excel(writer,'Comparativo Semanas')
df4.to_excel(writer,'RTs 40 Horas de Conexion')
df5.to_excel(writer,'RTs Criticos')
df_data.to_excel(writer, 'Estadisticas')

writer.save()
writer.close()
```
