```python
df_bonus25 = df_bonus25.fillna(0)
df_bonus60 = df_bonus60.fillna(0)
df_conexion = df_conexion.fillna(0)
```

```python
def getMonto(pedidos,hora,dia):
    if pedidos >= 60:
        x = df_bonus60.iloc[df_bonus60['Inicio'][hora].hour,dia]                            
        return x
    elif pedidos >= 25:
        x = df_bonus25.iloc[df_bonus25['Inicio'][hora].hour,dia]
        return x
    else:
        return 0
```

```python
def getMontoDia(pedidos):
    for i in df_vBonus.index:
        if pedidos>= df_vBonus['Viajes'][i]:
            bono = df_vBonus['Domingo 16'][i]            
    return bono
```
