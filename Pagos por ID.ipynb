```python
import pandas as pd
import numpy as np
import os
```

```python
mainpath = "C:/Users/mromo/Desktop/Rappi/Datasets/"
dep1 = {}
dep2 = {}
```

```python
for x in range(31,35):
    pagos = "S" + str(x) + "/" + "Pagos Vigentes.xlsx"
    pagable = os.path.join(mainpath,pagos)
    pag1 = 'Dep1'
    pag2 = 'Dep2'
    dep1[x] = pd.read_excel(pagable, pag1)
    dep2[x] = pd.read_excel(pagable, pag2)
    print("Carga semana %d ok"%(x))
```

```python
for x in range(31,35):
    dep1[x] = dep1[x].loc[:,['Estado','SKID','Monto Transferencia']]
    dep2[x] = dep2[x].loc[:,['Estado','SKID','Monto Transferencia']]
```

```python
dep1[31].columns.values
```

```python
pagado = pd.DataFrame(columns = dep1[31].columns)
pendiente = pd.DataFrame(columns = dep1[31].columns)
rechazado = pd.DataFrame(columns = dep1[31].columns)
```

```python
for x in range(31,35):
    pagado = pagado.append(dep1[x].loc[dep1[x]['Estado'] == "PAGADO"])
    pendiente = pendiente.append(dep1[x].loc[dep1[x]['Estado'] == "PENDIENTE"])
    rechazado = rechazado.append(dep1[x].loc[dep1[x]['Estado'] == "RECHAZADO"])
```

```python
pagado
```

```python
pagado = pagado.groupby(['SKID','Estado']).sum().reset_index()
pagado
```

```python
pagado.dtypes
```

```python
pendiente = pendiente.groupby(['SKID','Estado']).sum().reset_index()
```

```python
pendiente.dtypes
```

```python
rechazado = rechazado.groupby(['SKID','Estado']).sum().reset_index()
```

```python
rechazado
```

```python
consolidado = pd.merge(pagado,pendiente, how ='outer', on=['SKID'])
consolidado.dtypes
```

```python
consolidado = pd.merge(consolidado, rechazado, how='left', on=['SKID'])
```

```python
consolidado.fillna('')
```

```python
consolidado = consolidado.sort_values('SKID')
```
