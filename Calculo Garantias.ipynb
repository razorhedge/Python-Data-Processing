```python
import pandas as pd
import numpy as np
import os
import datetime
```

```python
semana = '37'
```

```python
mainpath = "C:/Users/mromo/Desktop/Rappi/Datasets/S"+semana+"/Data"
conexion_path = "Info Conexion.csv"
incentivos_path = "Incentivos Sem "+semana+".xlsx"
ganancias_path = "Detalle Ganancia S"+semana+".xlsx"
fp_earn = os.path.join(mainpath,ganancias_path)
fp_incentivos = os.path.join(mainpath,incentivos_path)
fp_conexion = os.path.join(mainpath,conexion_path)
```

```python
df_conexion = pd.read_csv(fp_conexion, parse_dates = ['fecha'])
df_earn = pd.read_excel(fp_earn, parse_dates = ['created_at'])
df_bonus25 = pd.read_excel(fp_incentivos, sheet_name = 'P25')
df_bonus60 = pd.read_excel(fp_incentivos, sheet_name = 'P60')
df_vBonus = pd.read_excel(fp_incentivos, sheet_name = 'Incentivos por Dia')
```

```python
df_conexion = df_conexion.fillna(0)
df_conexion = df_conexion.rename(columns = {'skid':'storekeeper_id'})
```

```python
df_conexion.head()
```

```python
df1 = df_conexion
```

```python
df_garantias = df_earn[df_earn['name']=='service_cost']
```

```python
df_earn = df_earn.sort_values(['storekeeper_id','created_at'],ascending = True)
df_earn['total'] = df_earn['total'].astype(int)
```

```python
lunvie = ['2018-09-10','2018-09-11','2018-09-12','2018-09-13','2018-09-14']
sadom = ['2018-09-15', '2018-09-16']
jueviedom = ['2018-09-13','2018-09-14','2018-09-16']
```

```python
df_ppay = df_garantias.groupby('storekeeper_id').total.count()
df_ppay2 = pd.DataFrame(df_ppay)
df_ppay2 = df_ppay2.reset_index()
df_ppay2 = df_ppay2.fillna(0)
```

```python
df_ppay2.head()
```

```python
df_garantias
```

```python
df_jueves= df_garantias[df_garantias['fecha']==13]
df_viernes= df_garantias[df_garantias['fecha']==14]
df_domingo= df_garantias[df_garantias['fecha']==16]
```

```python
df_jueves = pd.DataFrame(df_jueves.groupby('storekeeper_id').name.count())
df_viernes = pd.DataFrame(df_viernes.groupby('storekeeper_id').name.count())
df_domingo = pd.DataFrame(df_domingo.groupby('storekeeper_id').name.count())
```

```python
df_dias = pd.merge(df_jueves,df_viernes, how = 'outer', on = 'storekeeper_id')
df_dias = df_dias.merge(df_domingo, how = 'outer', on ='storekeeper_id')
df_dias = df_dias.rename(columns = {'name_x':'Jueves 13','name_y':'Viernes 14','name':'Domingo 16'})
df_dias = df_dias.fillna(0)
```

```python
df_semanaAM = df1.loc[df1['fecha'].isin(lunvie)]
df_semanaAM = df_semanaAM[df_semanaAM['storekeeper_id']!=1]
df_fdsAM = df1.loc[df1['fecha'].isin(sadom)]
df_fdsAM = df_fdsAM[df_fdsAM['storekeeper_id']!=1]
```

```python
df_semanaAM = df_semanaAM[df_semanaAM['hora']>=8]
df_semanaAM = df_semanaAM[df_semanaAM['hora']<=11]
df_fdsAM = df_fdsAM[df_fdsAM['hora']>=9]
df_fdsAM = df_fdsAM[df_fdsAM['hora']<=12]
```

```python
df_fdsAM.head(20)
```

```python
df_garantias = pd.merge(df_conexion,df_ppay2, how='left', on ='storekeeper_id')
df_garantias = df_garantias.fillna(0)
df_garantias = df_garantias.rename(columns = {'total':'Total Pedidos'})
```

```python
df_garantias.insert(8,'Garantia',0)
df_garantias.insert(9,'Garantia Efectiva',0)
df_garantias.insert(10,'Garantia Jueves',0)
df_garantias.insert(11, 'Garantia Viernes',0)
df_garantias.insert(12, 'Garantia Domingo',0)
df_garantias.insert(13, 'Garantia AM Semana',0)
df_garantias.insert(14, 'Garantia AM FDS',0)
```

```python
df_garantias = df_garantias[df_garantias['storekeeper_id']!=1]
```

```python
writer = pd.ExcelWriter('C:/Users/mromo/Desktop/Rappi/Datasets/S'+semana+'/Garantias S'+semana+' Final.xlsx')
df_garantias.to_excel(writer,'Garantias S'+semana)
df_bonus25.to_excel(writer,'P25')
df_bonus60.to_excel(writer,'P60')
df_vBonus.to_excel(writer,'Incentivos por Dia')
df_dias.to_excel(writer,'Pedidos por Dia')
df_semanaAM.to_excel(writer,'Semana AM')
df_fdsAM.to_excel(writer,'FDS AM')


writer.save()
writer.close()
```
